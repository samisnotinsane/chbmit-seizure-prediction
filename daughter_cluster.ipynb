```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
n_interictal_points = 1000 # no. of points in normal class
n_ictal_points = 250 # no. of points in seizure class

delta_interictal = np.random.normal(loc=3.12, scale=0.57, size=(n_interictal_points))
delta_ictal = np.random.normal(loc=2.06, scale=0.53, size=(n_ictal_points))

gamma_interictal = np.random.normal(loc=3.17, scale=0.22, size=(n_interictal_points))
gamma_ictal = np.random.normal(loc=4.05, scale=0.15, size=(n_ictal_points))
```

```python
# plot delta waves
fig, ax = plt.subplots()
ax.scatter(delta_interictal, gamma_interictal, label='Interictal', alpha=0.6)
ax.scatter(delta_ictal, gamma_ictal, label='Ictal', alpha=0.6)
ax.set_xlabel('Relative Bandpower (Delta)')
ax.set_ylabel('Relative Bandpower (Gamma)')
ax.legend(loc='best')
plt.show()
```
