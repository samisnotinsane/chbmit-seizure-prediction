```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

N = 400
ex = np.random.randn(N)
a = np.array([1.2])
S = ex.copy()
for n in range(1, N):
    x = np.array([S[n-1]])
    S[n] = x@a + ex[n]

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(6,4))

ax[0].plot(range(N), ex)
ax[0].set_title('Excitation Signal')

ax[1].plot(range(N), S)
ax[1].set_title('AR(1)')

fig.tight_layout()
```

```python
np.random.seed(42)

N = 400
ex = np.random.randn(N)
a = np.array([1.2, -0.4])
S = ex.copy()
for n in range(2, N):
    x = np.array([S[n-1], S[n-2]])
    S[n] = x@a + ex[n]

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(6,4), sharex=True)

ax[0].plot(range(N), ex, label='Observation')
ax[0].set_ylim([-5, 5])
ax[0].legend()
# ax[0].set_title('Excitation Signal')

ax[1].plot(range(N), S, label='Modelled')
ax[1].set_ylim([-8, 9])
ax[1].legend(loc=1)
ax[1].set_xlabel('Index, $n$')
# ax[1].set_title('AR(2)')

fig.tight_layout()
plt.savefig('../figures/autoregression/noise_modelling.pdf')
plt.show()
```

```python
np.random.seed(42)

N = 400
ex = np.random.randn(N)
a = np.array([1.2, -0.4, 0.1])
S = ex.copy()
for n in range(3, N):
    x = np.array([S[n-1], S[n-2], S[n-3]])
    S[n] = x@a + ex[n]

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(6,4))

ax[0].plot(range(N), ex)
ax[0].set_title('Excitation Signal')

ax[1].plot(range(N), S)
ax[1].set_title('AR(3)')

fig.tight_layout()
plt.show()
```
