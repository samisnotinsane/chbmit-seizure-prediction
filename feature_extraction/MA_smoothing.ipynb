# MA Smoothing

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 60
c = 60*np.ones(m)
c = c/c.sum()
print(c)
```

```python
n_samples = 1798
ar = np.random.randn(n_samples)
```

```python
ar_ma = np.zeros(n_samples)
for k in range(n_samples):
    for j in range(m):
        ar_ma[k] = c[j]*ar[k-j]
```

```python
ar_ma.shape
```

```python
fig, ax = plt.subplots(figsize=(12,3))
ax.plot(np.arange(0, n_samples), ar, color='tab:blue', label=r'$\hat{a}_k$') # raw
ax.plot(np.arange(0, n_samples), ar_ma, color='tab:red', label=r'$a_k$') # MA smoothed
ax.set_ylim([ar.min()*1.5, ar.max()*1.5])
ax.legend(loc=1)
ax.set_title('MA Smoothing of Signal', fontsize=16)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(figsize=(12,6), nrows=2)
axes[0].plot(np.arange(0, n_samples), ar, label=r'$\hat{a}_k$')
axes[0].set_ylim([ar.min()*1.5, ar.max()*1.5])
axes[0].legend(loc=1)
axes[0].set_title('Input Signal', fontsize=16) # input signal

axes[1].plot(np.arange(0, n_samples), ar_ma, color='tab:red', label=r'$a_k$')
axes[1].legend(loc=1)
axes[1].set_ylim([-2,2])
axes[1].set_title('MA Smoothed', fontsize=16) # MA smoothed signal

plt.tight_layout()
plt.show()
```
