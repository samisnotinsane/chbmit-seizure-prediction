```python
import classification.dataparser as dp
import pandas as pd
```

```python
rootdir = '/Volumes/My Passport/AI_Research/data/physionet.org/files/chbmit/1.0.0/'
path = './data_summary_copy.xlsx'
```

```python
for i in range(1, 10):
    patient_id = f'chb0{i}'
    parser = dp.DataParser(rootdir, patient_id)
    dict_files = parser._patient_summary()
    df = pd.DataFrame(data=dict_files)
    df = df.T
    with pd.ExcelWriter(path, engine='openpyxl', mode='a') as writer:
        df.to_excel(writer, sheet_name=f'chb{i}')
```
