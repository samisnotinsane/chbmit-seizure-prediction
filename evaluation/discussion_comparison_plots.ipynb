```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read 2nd sheet of an excel file
df_ARMA = pd.read_excel('discussion_plot.xlsx', sheet_name = 0)
df_Spectral = pd.read_excel('discussion_plot.xlsx', sheet_name = 1)
```

```python
print(df_ARMA.T[0])
```

```python
print(df_Spectral)
```

```python
g = sns.catplot(
    data=df_ARMA, kind="bar",
    x="TPR", y="FPR", hue="Algorithm",
    ci="sd", palette="Set2", alpha=.6, height=6
)
g.despine(left=True)
g.set_axis_labels("Preictal True Positive Rate (TPR)", "Interictal False Positive Rate (FPR)")
g.legend.set_title("")
# plt.legend(loc='upper left')
# plt.tight_layout()
plt.savefig('./figures/ARMA_Random_Comparison.pdf', bbox_inches = "tight")
```

```python
g = sns.catplot(
    data=df_Spectral, kind="bar",
    x="TPR", y="FPR", hue="Algorithm",
    ci="sd", palette="Set2", alpha=.6, height=6
)
g.despine(left=True)
g.set_axis_labels("Preictal True Positive Rate (TPR)", "Interictal False Positive Rate (FPR)")
g.legend.set_title("")
# plt.tight_layout()
plt.savefig('./figures/Spectral_Random_Comparison.pdf', bbox_inches = "tight")
```
