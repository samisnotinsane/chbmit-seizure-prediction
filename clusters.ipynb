# Data Clusters

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
n_cls_a = 600 # points in cluster a
n_cls_b = 1200 # points in cluster b

# cluster a
mu, sigma = 1, 0.5 # mean and standard deviation
sig_one = np.random.normal(mu, sigma, (n_cls_a,))

mu, sigma = 3, 0.3
sig_two = np.random.normal(mu, sigma, (n_cls_a,))

# cluster b
mu, sigma = -2, 1.5
sig_three = np.random.normal(mu, sigma, (n_cls_b,))

mu, sigma = -1, 0.7
sig_four = np.random.normal(mu, sigma, (n_cls_b,))
```

```python
fig, ax = plt.subplots()
ax.scatter(sig_one, sig_two, label='Cluster A / Anomaly', alpha=.6)
ax.scatter(sig_three, sig_four, label='Cluster B / Nominal', alpha=.6)
ax.legend(loc='best')
```

## Distribution

```python
plt.hist(sig_one, label='sig_one', color='tab:red', alpha=.8)
plt.hist(sig_two, label='sig_two', color='tab:cyan', alpha=.8)
plt.title('Frequency Histogram: Anomaly Cluster')
plt.xlabel('Unit')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
plt.hist(sig_three, label='sig_three', color='tab:red', alpha=.8)
plt.hist(sig_four, label='sig_four', color='tab:cyan', alpha=.8)
plt.title('Frequency Histogram: Nominal Cluster')
plt.xlabel('Unit')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
